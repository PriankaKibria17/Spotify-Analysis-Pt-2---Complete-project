2. Create filters to find out:

     - a. How many athletes who participated in the Sport "Tug-Of-War" were from Team "Sweden" in our sample dataset?

     - b. How many Gold medals were awarded to athletes who were over 40 years old?



```python
# a. How many athletes who participated in the Sport "Tug-Of-War" were from Team "Sweden" in our sample dataset?
Sport_Filter = Olympian_df['Sport'] == "Tug-Of-War"
Team_Filter = Olympian_df['Team'] == "Sweden"
Sport_Team = Olympian_df.loc[Sport_Filter & Team_Filter]
Sport_Team
```

```python
# using len to obtain amount of participants in numeric value
Participants = len(Sport_Team)
print('The amount of athletes participating in tug-of-war from Sweden is', Participants)
```

```python
# b. How many Gold medals were awarded to athletes who were over 40 years old?
# using 2 conditions for age those who are over 40 and medals equivalent to gold only
#combining my filters and naming it "Medals_earned"
Medalist_Over40 = Olympian_df['Age'] > 40
Gold = Olympian_df['Medal'] == 'Gold'
Medals_earned = Olympian_df.loc[Medalist_Over40 & Gold]
Medals_earned
```

```python
# obtaining the amount of athletes who were over 40 and awarded gold medals
gold_amount = len(Medals_earned)
print('The amount of athletes over 40 who earned gold medals is', gold_amount )
```
